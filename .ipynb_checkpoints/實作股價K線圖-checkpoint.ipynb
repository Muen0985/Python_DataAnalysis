# 實作股價K線圖
台積電 TSMC (2330) <br>
聯發科 MediaTek (2454) 

```python
import yfinance as yf
import pandas as pd
import datetime as dt

import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
start=dt.datetime(2023,6,1)
end=dt.datetime(2023,7,17)
```

```python
import yfinance as yf

# 取得台積電的股價
tsmc = yf.download('2330.TW', start, end)

# 取得聯發科的股價
med = yf.download('2454.TW', start, end)

print(tsmc.head())
print(med.head())
```

```python
tsmc
```

```python
med
```

```python
plt.rcParams['font.family'] = 'Microsoft JhengHei'
```

### 台積電 VS 聯發科 股價線圖

```python
fig=plt.figure(figsize=(10,6))
plt.title('台積電 VS 聯發科 股價線圖',loc='center')
plt.plot(tsmc['Close'],label='TSMC')
plt.plot(med['Close'],label='Mediatec')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True,axis='y')
plt.legend()
plt.show()
```
