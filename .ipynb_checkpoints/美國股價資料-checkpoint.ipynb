```python
import pandas as pd
import datetime as dt
import pandas_datareader.data as data
```

```python
start=dt.datetime(2023,1,1)
print(f'start: {start}')
end=dt.datetime(2023,6,2)
df=data.DataReader('AAPL','stooq',start,end)
df
```

### DatetimeIndex
以時間作為索引的型態

```python
df.index
```

```python
df.index.isocalendar().head()
# isocalendar: 回傳一個 tuple，內容分別是 ( 年、第幾週、isoweekday )
# isoweekday: 回傳一星期中的第幾天，星期一為 1
```

```python
df['week']=df.index.isocalendar().week
```

```python
df
```

```python
df.groupby('week').Volume.sum()
```

## resample
調整週期重新採樣的方法<br>
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.resample.html#pandas.DataFrame.resample

#### 可以用來調整周期，重新採樣。 例如分析股票十日均線 (將十日內的股價加總起來平均)

```python
ten=df.resample('10D').mean() # 代表10天的平均
```

```python
ten
```

```python
mon=df.resample('1M').mean()
mon
```
