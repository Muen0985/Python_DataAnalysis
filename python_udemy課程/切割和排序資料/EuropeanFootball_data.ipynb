### Series

```python
euro['Goals'] # 是顯示series物件
```

```python
round( euro['Goals'].mean() ) #做 mean 的四捨五入
```

```python
euro['Goals']>5
```

```python
euro[euro['Goals']>5] #加入條件
```

###  Series 中專門處理字串的函式 (Series物件.str.)

```python
euro['Team'].str.startswith('S') # 產生出過濾條件
#series物件  #函式 
```

```python
euro[euro['Team'].str.startswith('S')]
```

### 切割方式取值

```python
# 用Pandas的 iloc 來做切割
```

```python
euro.iloc[0:3,0:5] # 前面的是列的數量，後面是欄的數量
```

```python
# 用行列名稱做切割
```

```python
euro.loc['0':'3','Team':'Blocks']  # 皆是用欄或列的名稱當參數
```
