###### 將 item_price 欄位型態從「字串」轉為浮點數Float型態¶

```python
for i in chipo['item_price'][:5]:  # 取前5個當範例
    print(i[1:])
```

```python
price=[float(value[1:]) for value in chipo['item_price']]
chipo['item_price']=price
```

```python
chipo['item_price'].mean()
```
