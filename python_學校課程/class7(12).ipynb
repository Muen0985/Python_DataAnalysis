```python
# body > div.main-info.article-main-info > div > div > div > div.text-muted.article-info > span:nth-child(2) > time
```

```python
contents=[]
time=[]

for href in hrefs:
    resp=requests.request(method='GET',url=href)
    soup=BeautifulSoup(markup=resp.text,features='html.parser')
    
    nodes=soup.select('div > article')
    contents.append(''.join([node.text.replace("\n","") for node in nodes]))
    nodes=soup.select('span:nth-child(2) > time')
    time.append(''.join([node.text for node in nodes]))
```

```python
url = "https://www.drmaster.com.tw/Publish_Search.asp"
payloads={'selectID': '3','content': '陳祥輝'}
resp=requests.request(method='POST',url=url,data=payloads)
```

Pandas套件提供了read_html()方法(Method)，只要傳入網址，就能夠讀取網頁中的table表格標籤，並且回傳一個Pandas DataFrame物件。

```python
resp.encoding='utf-8'
result=pd.read_html(resp.text,header=0)
for idx,df in enumerate(result,start=0):
    print(f"{idx} -------------------")
    print(df)
```
